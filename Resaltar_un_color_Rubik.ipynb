```
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```
lower_blue=np.array([5,160,100])
upper_blue=np.array([20, 255, 255])
```

```
img=cv2.imread("claro.png")

img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_hsv=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)

mask=cv2.inRange(img_hsv,lower_blue,upper_blue)

mask=cv2.GaussianBlur(mask,(3,3),0)
result=cv2.bitwise_and(img,img,mask=mask)

plt.figure(figsize=(6,4))
plt.imshow(result)
plt.show()
```

```
img2=cv2.imread("oscuro.png")

img2=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img_hsv2=cv2.cvtColor(img2,cv2.COLOR_RGB2HSV)

mask2=cv2.inRange(img_hsv2,lower_blue,upper_blue)

mask2=cv2.GaussianBlur(mask2,(3,3),0)
result2=cv2.bitwise_and(img2,img2,mask=mask2)

plt.figure(figsize=(6,4))
plt.imshow(result2)
plt.show()
```
