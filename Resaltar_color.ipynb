```
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```
pez =cv2.imread("nemo0.jpg")


pez = cv2.cvtColor(pez,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.subplot(1,2,1),plt.imshow(pez)

plt.show()
```

```
orange = np.uint8([[[239, 127, 26]]])
hsv_orange = cv2.cvtColor(orange,cv2.COLOR_RGB2HSV)
print(hsv_orange)
```

```
pez2 = cv2.cvtColor(pez,cv2.COLOR_RGB2HSV)

lower_orange = np.array([1,170,130])
upper_orange = np.array([20,255,255])

mask = cv2.inRange(pez2,lower_orange,upper_orange)


print(mask.dtype)
print(np.max(mask))
print(np.min(mask))


plt.imshow(mask,cmap = 'gray')

plt.show()
```

```
img = np.zeros((pez.shape),dtype = np.uint8)
print(img[100][180])
counti=0
for i in mask:
  countj=0
  for j in i:
    if(j==255):
      img[counti][countj]=pez[counti][countj]
    countj=countj+1
  counti=counti+1

plt.imshow(img)

plt.show() 
```
