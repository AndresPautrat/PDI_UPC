```
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.zeros((256,256,3),dtype = np.uint8)

r=255
g=255
b=255

img[0:64,:,:] = [255,0,0]
img[0:32,:,:] = [r,g,b]


img[64:128,:,:] = [0,255,0]

img[128:196,:,:] = [0,0,255]
img[196:256,:,:] = [255,255,255]

plt.imshow(img)
plt.show()
```

```
img[0,:,:] = [255,255,0]
for i in range(256):
  for j in range(256):
    img[i,j,:] = [255,255-i,j]
 
plt.imshow(img)
plt.show()

#h = angulo = combinaciones
#s = distancia = tono claro 
#v = profundidad = tono oscuro
```

```
img2 = np.zeros((360,256,3),np.uint8)
for interval in range(180):
  a1 = interval*2
  a2 = (interval+1)*2
  img2[a1:a2,:,:] = [interval,255,255]

outImg = cv2.cvtColor(img2,cv2.COLOR_HSV2RGB)

plt.imshow(outImg)
plt.show()
```

```
img=cv2.imread("pathology_cll20x01.jpg")

img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
lower_blue=np.array([110,250,0])
upper_blue=np.array([130, 255, 255])

mask=cv2.inRange(img_hsv,lower_blue,upper_blue)
print(mask.dtype)
print(np.max(mask))
print(np.min(mask))
plt.imshow(mask, cmap='gray')
plt.show()
```

```
plt.imshow(mask, cmap='gray')
plt.show()
```

```
mask=cv2.GaussianBlur(mask,(3,3),0)
result=cv2.bitwise_and(img,img,mask=mask)

plt.figure(figsize=(6,4))
plt.imshow(result)
plt.show()


lower_blue=np.array([0,150,110])
upper_blue=np.array([25, 255, 240])
```

```
img=cv2.imread("claro.png")
img2=cv2.imread("oscuro.png")

img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img_hsv=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
img_hsv2=cv2.cvtColor(img2,cv2.COLOR_RGB2HSV)
lower_blue=np.array([5,160,100])
upper_blue=np.array([20, 255, 255])

mask=cv2.inRange(img_hsv,lower_blue,upper_blue)
mask2=cv2.inRange(img_hsv2,lower_blue,upper_blue)
plt.imshow(mask, cmap='gray')
plt.show()
plt.imshow(mask2, cmap='gray')
plt.show()

mask=cv2.GaussianBlur(mask,(3,3),0)
result=cv2.bitwise_and(img,img,mask=mask)

plt.figure(figsize=(6,4))
plt.imshow(result)
plt.show()
```
